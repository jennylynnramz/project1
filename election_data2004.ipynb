```python
election_data_all= pd.read_csv(election_data_all_csv)
election_data_all
```

```python
election_data_all_dropNA = election_data_all.dropna()
```

```python
election_data_2004_1 = election_data_all_dropNA[election_data_all_dropNA['year']==2004]
election_data_2004_1
```

```python
election_data_2004_2 = election_data_2004_1.drop(columns=['FIPS', 'version', 'office'])
election_data_2004_2
```

```python
election_data_2004_3 = election_data_2004_2.reset_index(drop=True)
```

```python
election_data_2004_3.columns=["year", "state", "state_code", "county", "candidate", "party", "candidate_votes", "total_votes"]
election_data_2004_3
```
