```python
from census import Census
from us import states
import requests
import json
from pprint import pprint
from api import APIkey
c = Census(APIkey, year=2000)
#reference = https://api.census.gov/data.html

url = "https://api.census.gov/data/2000/sf3"

#url = "https://api.census.gov/data/2014/pep/natstprc?get=STNAME,POP&DATE_=7&for=state:*"
#for all states (&for=state:*)
####Please note:
#####Predicates always start with an ampersand (&).
#####A wildcard (:*) can be included to search for all the values of a variable. Wildcards work for geographies and string variables only.
#####A variable should appear only once in a query Ì¶ either in the get statement or as a predicate.
######In the example above, DATE_ is a predicate (&DATE_=7), so you do not need to include it in the get statement (?get=STATENAME,POP).
```
