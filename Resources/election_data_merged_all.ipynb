```python
#import dependencies 
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import sem
from scipy.stats import linregress
from pprint import pprint
```

```python
#file locations
election_data_csv_2000 = "Resources/FOR_JINHO_2000.csv"
election_data_csv_2004 = "Resources/FOR_JINHO_2004.csv"
election_data_csv_2008 = "Resources/FOR_JINHO_2008.csv"
election_data_csv_2012 = "Resources/FOR_JINHO_2012.csv"
election_data_csv_2016 = "Resources/FOR_JINHO_2016.csv"
```

```python
#read the files
elect_2000= pd.read_csv(election_data_csv_2000)
elect_2004= pd.read_csv(election_data_csv_2004)
elect_2008= pd.read_csv(election_data_csv_2008)
elect_2012= pd.read_csv(election_data_csv_2012)
elect_2016= pd.read_csv(election_data_csv_2016)
```

```python
elect_2000 = elect_2000.rename(columns = {'total_votes':'2000_total_votes'})
```

```python
elect_2004
```

```python
elect_2008 = elect_2008.rename(columns = {'total_votes':'2008_total_votes'})
```

```python
elect_2012 = elect_2012.rename(columns = {'total_votes':'2012_total_votes'})
```

```python
elect_2016
```

```python
#correcting formatting on 2004 df
el_df_2004 = elect_2004.drop(['County','State','State_Code','Year','County_Pop_2004','Candidate'], axis=1)
el_df_2004
```

```python
# create df for unique parties 
democrat_2004 = el_df_2004.loc[el_df_2004["Party"] == 'democrat']
democrat_2004
```

```python
republican_2004 = el_df_2004.loc[el_df_2004["Party"] == 'republican']
republican_2004
```

```python
republican_2004 = republican_2004[['Total_Votes', 'Candidate_Votes', 'Location']]
republican_2004.columns = ['total_votes', '2004_republican_votes', 'Location']

democrat_2004 = democrat_2004[['Total_Votes', 'Candidate_Votes', 'Location']]
democrat_2004.columns = ['total_votes', '2004_democrat_votes', 'Location']
```

```python
print(len(republican_2004['Location'].unique()), len(democrat_2004['Location'].unique()))
```

```python
election_2004_clean = pd.merge(republican_2004,democrat_2004, on='Location')
election_2004_clean = election_2004_clean[['Location','total_votes_x','2004_republican_votes','2004_democrat_votes']]
election_2004_clean.columns = ['Location','total_votes','2004_republican_votes','2004_democrat_votes']
election_2004_clean
```

```python
election_2004_clean = election_2004_clean.rename(columns = {'total_votes':'2004_total_votes'})
```

```python
#do the same for 2016
el_df_2016 = elect_2016.drop(['County','State','State_Code','Year','County_Pop_2016','Candidate'], axis=1)
el_df_2016
```

```python
# create df for unique parties 
democrat_2016 = el_df_2016.loc[el_df_2016["Party"] == 'democrat']
democrat_2016
```

```python
republican_2016 = el_df_2016.loc[el_df_2016["Party"] == 'republican']
republican_2016
```

```python
republican_2016 = republican_2016[['Total_Votes', 'Candidate_Votes', 'Location']]
republican_2016.columns = ['total_votes', '2016_republican_votes', 'Location']

democrat_2016 = democrat_2016[['Total_Votes', 'Candidate_Votes', 'Location']]
democrat_2016.columns = ['total_votes', '2016_democrat_votes', 'Location']
```

```python
print(len(republican_2004['Location'].unique()), len(democrat_2004['Location'].unique()))
```

```python
election_2016_clean = pd.merge(republican_2016,democrat_2016, on='Location')
election_2016_clean = election_2016_clean[['Location','total_votes_x','2016_republican_votes','2016_democrat_votes']]
election_2016_clean.columns = ['Location','total_votes','2016_republican_votes','2016_democrat_votes']
election_2016_clean
```

```python
election_2016_clean = election_2016_clean.rename(columns = {'total_votes':'2016_total_votes'})
```

```python
#merge 2000 and 2004
merge_df1 = pd.merge(elect_2000, election_2004_clean, on="Location", how="inner")
merge_df1
```

```python
#merge 2008
merge_df2 = pd.merge(merge_df1, elect_2008, on="Location", how="inner")
merge_df2
```

```python
#merge 2012
merge_df3 = pd.merge(merge_df2, elect_2012, on="Location", how="inner")
merge_df3
```

```python
#merge 2016
merge_df4 = pd.merge(merge_df3, election_2016_clean, on="Location", how="inner")
merge_df4
```

```python
merge_df4.to_csv(r'Resources/election_2000_2016_merged.csv', index=False, header=True)
```
